```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
X = np.array([[1],[2.1],[3]])
y = np.array([[1],[2],[3]])
x_pred = np.array([[4]])
```

```python
model = LinearRegression()
model.fit(X, y)
prediction = model.predict(x_pred)
```

```python
print("Array:", prediction)
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("Score:", model.score(X, y))
```

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
X, y, coef = make_regression(
    n_samples=100,
    n_features=1,
    n_informative=1,
    noise=10,
    coef=True,
    random_state=0,
    bias=100.0
)
```

```python
plt.scatter(X, y)
plt.title("X (1 Dimension) and y")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
print(f"Equation: y = {model.coef_[0]:.8f} * x + {model.intercept_:.8f}")
```

```python
plt.scatter(X, y, label="Data")
plt.plot(X, model.predict(X), color='red', label="Fitted Line")
plt.title("Data and Fitted Line (noise=10)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
y_pred = model.predict(X)
print("First 10 predictions:", y_pred[:10])
```

```python
def compute_mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

mse = compute_mse(y, y_pred)
print("MSE:", mse)
```

```python
X2, y2, coef2 = make_regression(
    n_samples=100,
    n_features=1,
    n_informative=1,
    noise=50,
    coef=True,
    random_state=0,
    bias=100.0
)
```

```python
model2 = LinearRegression()
model2.fit(X2, y2)
y_pred2 = model2.predict(X2)
```

```python
plt.scatter(X2, y2, label="Data")
plt.plot(X2, y_pred2, color='red', label="Fitted Line")
plt.title("Data and Fitted Line (noise=50)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
mse2 = compute_mse(y2, y_pred2)
print("MSE (noise=50):", mse2)
```

```python
from sklearn.datasets import load_diabetes
import pandas as pd
```

```python
diabetes = load_diabetes(as_frame=True)
X, y = diabetes.data, diabetes.target
df = pd.concat([X, y.rename("target")], axis=1)
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
pd.set_option('display.float_format', '{:.5f}'.format)
print(df.describe())
```

```python
print(df.info())
```

```python
df['age_bin'] = pd.cut(df['age'], bins=3, labels=['young', 'middle', 'old'])
df['bmi_bin'] = pd.cut(df['bmi'], bins=3, labels=['low', 'medium', 'high'])
```

```python
freq_table = pd.crosstab(df['age_bin'], df['bmi_bin'])
print(freq_table)
```

```python
pivot = df.pivot_table(values='target', index='age_bin', columns='bmi_bin', aggfunc='mean')
print(pivot)
```

```python
grouped = df.groupby(['age_bin', 'bmi_bin'])['target'].mean()
print(grouped)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=43
)
```

```python
print(y_train.values[:10])
```

```python
print(y_test.values[:10])
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

for name, coef in zip(X_train.columns, model.coef_):
    print(f"{name}: {coef}")
print("Intercept:", model.intercept_)
```

```python
y_pred_test = model.predict(X_test)
y_pred_train = model.predict(X_train)
```

```python
print(y_pred_test[:10].reshape(-1, 1))
```

```python
from sklearn.metrics import mean_squared_error

mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)

print("Train MSE:", mse_train)
print("Test MSE:", mse_test)
```
